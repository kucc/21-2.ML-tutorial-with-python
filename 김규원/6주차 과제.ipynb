```python
def month(num):
    if num == 1: return 0
    elif num == 2: return 1
    elif num == 3: return 2
    elif num == 4: return 3
    elif num == 5: return 4
    elif num == 6: return 5
    elif num == 7: return 6
    elif num == 8: return 7
    elif num == 9: return 8
    elif num == 10: return 9
    elif num == 11: return 10
    elif num == 12: return 11
```

```python
labeled_bike_df['month']= labeled_bike_df['month'].apply(month)
```

# data split -> 정규화한 데이터로 모두 사용했으므로 log변환 했던걸 되돌리는 작업은 안했습니다

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from scipy import stats
from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split
%matplotlib inline

#labeled
l_y_target = labeled_bike_df['count']
l_X_data = labeled_bike_df.drop(['count'],axis=1,inplace=False)

l_X_train , l_X_test , l_y_train , l_y_test = train_test_split(l_X_data , l_y_target ,test_size=0.3, random_state=156)

#one hot encoded
o_y_target = ohe_bike_df['count']
o_X_data = ohe_bike_df.drop(['count'],axis=1,inplace=False)

o_X_train , o_X_test , o_y_train , o_y_test = train_test_split(o_X_data , o_y_target ,test_size=0.3, random_state=156)
```

## label encoding data + Ridge

```python
# label encoding + Ridge
from sklearn.linear_model import Ridge
from sklearn.model_selection import cross_val_score

ridge = Ridge(alpha = 10)
neg_mse_scores = cross_val_score(ridge, l_X_train, l_y_train, scoring="neg_mean_squared_error", cv = 5)
rmse_scores  = np.sqrt(-1 * neg_mse_scores)
avg_rmse = np.mean(rmse_scores)
print(' 5 folds 의 개별 RMSE scores : ', np.round(rmse_scores,3))
print(' 5 folds 의 평균 RMSE : {0:.3f} '.format(avg_rmse))
```

## one hot encoding data + Ridge

```python
# one hot encoding + Ridge
from sklearn.linear_model import Ridge
from sklearn.model_selection import cross_val_score

ridge = Ridge(alpha = 10)
neg_mse_scores = cross_val_score(ridge, o_X_train, o_y_train, scoring="neg_mean_squared_error", cv = 5)
rmse_scores  = np.sqrt(-1 * neg_mse_scores)
avg_rmse = np.mean(rmse_scores)
print(' 5 folds 의 개별 RMSE scores : ', np.round(rmse_scores,3))
print(' 5 folds 의 평균 RMSE : {0:.3f} '.format(avg_rmse))
```

## one hot encoding이 더 좋으므로 밑부터 onehotencoded data로만 했습니다

# Ridge

```python
# alpha값 하이퍼파라미터튜닝
alphas = [0 , 0.1 , 1 , 10 , 100]

# alphas list 값을 iteration하면서 alpha에 따른 평균 rmse 구함.
for alpha in alphas :
    ridge = Ridge(alpha = alpha)
    
    #cross_val_score를 이용하여 5 fold의 평균 RMSE 계산
    neg_mse_scores = cross_val_score(ridge, o_X_train, o_y_train, scoring="neg_mean_squared_error", cv = 5)
    avg_rmse = np.mean(np.sqrt(-1 * neg_mse_scores))
    print('alpha {0} 일 때 5 folds 의 평균 RMSE : {1:.3f} '.format(alpha,avg_rmse))
```

```python
for alpha in alphas :
    ridge = Ridge(alpha = alpha)
    ridge.fit(o_X_train , o_y_train)
    
    coeff = pd.Series(data=ridge.coef_, index=o_X_train.columns)    
    print('alpha {0} 일 때 회귀계수 :'.format(alpha))
    print(coeff.sort_values(ascending=False))
    print()
```

# Lasso

```python
from sklearn.linear_model import Lasso

lasso = Lasso(alpha = 10)
neg_mse_scores = cross_val_score(lasso, o_X_train, o_y_train, scoring="neg_mean_squared_error", cv = 5)
rmse_scores  = np.sqrt(-1 * neg_mse_scores)
avg_rmse = np.mean(rmse_scores)
print(' 5 folds 의 개별 RMSE scores : ', np.round(rmse_scores,3))
print(' 5 folds 의 평균 RMSE : {0:.3f} '.format(avg_rmse))
```

```python
# alpha값 하이퍼파라미터튜닝
alphas = [0.07, 0.1, 0.5, 1, 3]

# alphas list 값을 iteration하면서 alpha에 따른 평균 rmse 구함.
for alpha in alphas :
    lasso = Lasso(alpha = alpha)
    
    #cross_val_score를 이용하여 5 fold의 평균 RMSE 계산
    neg_mse_scores = cross_val_score(lasso, o_X_train, o_y_train, scoring="neg_mean_squared_error", cv = 5)
    avg_rmse = np.mean(np.sqrt(-1 * neg_mse_scores))
    print('alpha {0} 일 때 5 folds 의 평균 RMSE : {1:.3f} '.format(alpha,avg_rmse))
```

```python
coeff_df = pd.DataFrame()
for alpha in alphas :
    lasso = Lasso(alpha = alpha)
    lasso.fit(o_X_train , o_y_train)
    
    coeff = pd.Series(data=lasso.coef_, index=o_X_train.columns)
    colname='alpha:'+str(alpha)
    coeff_df[colname] = coeff
    
sort_column = 'alpha:'+str(alphas[0])
coeff_df.sort_values(by=sort_column, ascending=False)
```

# ElasticNet

```python
from sklearn.linear_model import ElasticNet

alphas = [0.07, 0.1, 0.5, 1, 3]

# alphas list 값을 iteration하면서 alpha에 따른 평균 rmse 구함.
for alpha in alphas :
    elsnet = ElasticNet(alpha = alpha, l1_ratio=0.7)
    
    #cross_val_score를 이용하여 5 fold의 평균 RMSE 계산
    neg_mse_scores = cross_val_score(elsnet, o_X_train, o_y_train, scoring="neg_mean_squared_error", cv = 5)
    avg_rmse = np.mean(np.sqrt(-1 * neg_mse_scores))
    print('alpha {0} 일 때 5 folds 의 평균 RMSE : {1:.3f} '.format(alpha,avg_rmse))
```

```python
coeff_df = pd.DataFrame()
for alpha in alphas :
    elsnet = ElasticNet(alpha = alpha, l1_ratio=0.7)
    elsnet.fit(o_X_train , o_y_train)
    
    coeff = pd.Series(data=elsnet.coef_, index=o_X_train.columns)
    colname='alpha:'+str(alpha)
    coeff_df[colname] = coeff
    
sort_column = 'alpha:'+str(alphas[0])
coeff_df.sort_values(by=sort_column, ascending=False)
```

# 트리기반 회귀

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor

dt_reg = DecisionTreeRegressor(random_state=0, max_depth=4)
rf_reg = RandomForestRegressor(random_state=0, n_estimators=1000)
gb_reg = GradientBoostingRegressor(random_state=0, n_estimators=1000)
xgb_reg = XGBRegressor(n_estimators=1000)
lgb_reg = LGBMRegressor(n_estimators=1000)
```

```python
from sklearn.metrics import mean_squared_error

models = [dt_reg, rf_reg, gb_reg, xgb_reg, lgb_reg]

for model in models:
    model.fit(o_X_train, o_y_train)
    o_y_pred = model.predict(o_X_test)
    mse = mean_squared_error(o_y_pred, o_y_test)
    rmse = mse**(1/2)
    
    print(model.__class__.__name__,'회귀 모델의 rmse 값:', rmse)
```
